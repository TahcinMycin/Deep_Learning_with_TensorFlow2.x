```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data = pd.read_csv('data.csv')
```

```
X, y = data['Size'], data['Price']
```

```
X = np.array([4, 6, 8, 10, 14, 16]).reshape((-1, 1))
y = np.array([350, 775, 1150, 1200, 1395, 1675])
```

```
plt.scatter(X, y)
plt.grid(True)
plt.show()
```

```
import tensorflow as tf
```

```
deep_model = tf.keras.Sequential(
    tf.keras.layers.Dense(units=1, input_shape=[1])
)
```

```
deep_model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

```
deep_model.fit(X, y, epochs=1000, verbose=True)
```
